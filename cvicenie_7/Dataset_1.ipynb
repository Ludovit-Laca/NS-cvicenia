```python
from sklearn.datasets import load_iris
import numpy as np
```

```python
moj_IRIS_dataset = load_iris()
```

```python
moj_IRIS_dataset
```

```python
X = moj_IRIS_dataset.data[:, (0,3)]
```

```python
y = (moj_IRIS_dataset.target == 0).astype(np.int)
```

```python
y
```

```python
import pandas as pd
```

```python
moje_data = pd.DataFrame(data = np.c_[moj_IRIS_dataset['data'], moj_IRIS_dataset['target']], columns=moj_IRIS_dataset['feature_names']+['target'])
```

```python
moje_data
```

```python
# navrh dalsej ns - viacvrstvovy perceptron
from sklearn.neural_network import MLPClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_trenovacie, X_testovacie, y_trenovacie, y_testovacie = train_test_split(X,y,train_size=0.2)
```

```python
viacvrstvovy_perceptron = MLPClassifier(activation='tanh', alpha=100, hidden_layer_sizes=(5,2), max_iter=1000)
```

```python
viacvrstvovy_perceptron
```

```python
viacvrstvovy_perceptron.fit(X_trenovacie,y_trenovacie)
```

```python
viacvrstvovy_perceptron.score(X_testovacie, y_testovacie)
```
