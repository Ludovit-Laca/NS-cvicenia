0. Imports

```
import matplotlib.pyplot as plt
import numpy as np
```

1. Define **AND function**

```
# input layer
ax = np.array([[0,0],[0,1],[1,0],[1,1]])

# output layer 
ay = np.array([[0],[0],[0],[1]])
```

2. Define **OR function**

```
# input layer
ox = np.array([[0,0],[0,1],[1,0],[1,1]])

# output layer 
oy = np.array([[0],[1],[1],[1]])
```

3. Define **NOT function**

```
# input layer
nx = np.array([[0],[1]])

# output layer 
ny = np.array([[1],[0]])
```

4. Define activation function

```
# np.exp(1) is euler's number
def akt_function(x):
  return 1/(1 + np.exp(-x))
```

5. Define derivative function

```
# euler's number^x' - euler's number^x
def derivate_akt_function(x):
  return x * (1-x)
```

6. Sizes for **AND, OR, NOT** functions

```
# AND, OR functions
inputLayerSize = 2
hiddenLayerSize = 4
outputLayerSize = 1

# NOT function
not_inputLayerSize = 1
not_hiddenLayerSize = 1
not_outputLayerSize = 1
```

7. Define weights for **AND function**

```
# AND function weights
and_weight_hidden_layer = np.random.uniform(low=0.0, high=1.0, size=(inputLayerSize, hiddenLayerSize))
and_weight_output_layer = np.random.uniform(low=0.0, high=1.0, size=(hiddenLayerSize, outputLayerSize))

print("AND function hidden layer weight = " + str(and_weight_hidden_layer) + "\n")
print("AND function output layer weight = " + str(and_weight_output_layer) + "\n")
```

8. Define weights for **OR function**

```
# OR function weights
or_weight_hidden_layer = np.random.uniform(low=0.0, high=1.0, size=(inputLayerSize, hiddenLayerSize))
or_weight_output_layer = np.random.uniform(low=0.0, high=1.0, size=(hiddenLayerSize, outputLayerSize))

print("OR function hidden layer weight = " + str(or_weight_hidden_layer) + "\n")
print("OR function output layer weight = " + str(or_weight_output_layer) + "\n")
```

9. Define weights for **NOT function**

```
# NOT function weights
not_weight_hidden_layer = np.random.uniform(low=0.0, high=1.0, size=(not_inputLayerSize, not_hiddenLayerSize))
not_weight_output_layer = np.random.uniform(low=0.0, high=1.0, size=(not_hiddenLayerSize, not_outputLayerSize))

print("NOT function hidden layer weight = " + str(not_weight_hidden_layer) + "\n")
print("NOT function output layer weight = " + str(not_weight_output_layer) + "\n")
```

10. Neural network training - **AND function**

```
and_array_errors = []
and_array_output = []

for i in range(2000):
  and_hidden_layer = akt_function(np.matmul(ax, and_weight_hidden_layer))
  and_output_layer = akt_function(np.matmul(and_hidden_layer, and_weight_output_layer))

  and_error = ay - and_output_layer

  and_delta_Z = and_error*derivate_akt_function(and_output_layer)
  and_delta_H = and_delta_Z.dot(and_weight_output_layer.T)*derivate_akt_function(and_hidden_layer)

  and_weight_output_layer += and_hidden_layer.T.dot(and_delta_Z)
  and_weight_hidden_layer += ax.T.dot(and_delta_H)

  and_array_errors.append(float(and_error[3:]))
  and_array_output.append(float(and_weight_output_layer[3:]))


print("AND function output layer = " + str(and_output_layer) + "\n")
print("AND function error = " + str(and_error) + "\n")

print("AND function output layer weight = " + str(and_weight_output_layer) + "\n")
print("AND function hidden layer weight = " + str(and_weight_hidden_layer) + "\n")
```

11. Neural network training - **OR function**

```
or_array_errors = []
or_array_output = []

for i in range(2000):
  or_hidden_layer = akt_function(np.matmul(ox, or_weight_hidden_layer))
  or_output_layer = akt_function(np.matmul(or_hidden_layer, or_weight_output_layer))

  or_error = oy - or_output_layer

  or_delta_Z = or_error*derivate_akt_function(or_output_layer)
  or_delta_H = or_delta_Z.dot(or_weight_output_layer.T)*derivate_akt_function(or_hidden_layer)

  or_weight_output_layer += or_hidden_layer.T.dot(or_delta_Z)
  or_weight_hidden_layer += ox.T.dot(or_delta_H)

  or_array_errors.append(float(or_error[3:]))
  or_array_output.append(float(or_weight_output_layer[3:]))


print("OR function output layer = " + str(or_output_layer) + "\n")
print("OR function error = " + str(or_error) + "\n")

print("OR function output layer weight = " + str(or_weight_output_layer) + "\n")
print("OR function hidden layer weight = " + str(or_weight_hidden_layer) + "\n")
```

12. Neural network training - **NOT function**

```
not_array_errors = []
not_array_output = []

for i in range(2000):
  not_hidden_layer = akt_function(np.matmul(nx, not_weight_hidden_layer))
  not_output_layer = akt_function(np.matmul(not_hidden_layer, not_weight_output_layer))

  not_error = ny - not_output_layer

  not_delta_Z = not_error*derivate_akt_function(not_output_layer)
  not_delta_H = not_delta_Z.dot(not_weight_output_layer.T)*derivate_akt_function(not_hidden_layer)

  not_weight_output_layer += not_hidden_layer.T.dot(not_delta_Z)
  not_weight_hidden_layer += nx.T.dot(not_delta_H)

  not_array_errors.append(float(not_error[1:]))
  not_array_output.append(float(not_weight_output_layer[0:]))


print("NOT function output layer = " + str(not_output_layer) + "\n")
print("NOT function error = " + str(not_error) + "\n")

print("NOT function output layer weight = " + str(not_weight_output_layer) + "\n")
print("NOT function hidden layer weight = " + str(not_weight_hidden_layer) + "\n")
```

13. Show function errors in a plot

```
plt.title("Errors")
plt.plot(and_array_errors, label="AND")
plt.plot(or_array_errors, label="OR")
plt.plot(not_array_errors, label="NOT")
plt.legend()
```

14. Show function weights in a plot

```
plt.title("Weights")
plt.plot(and_array_output, label="AND")
plt.plot(or_array_output, label="OR")
plt.plot(not_array_output, label="NOT")
plt.legend()
```
