5.) Transformujte počet stĺpcov a riadkov matice

```
N = 7
M = 7
matica = np.random.randint(0,9, size=(N,M))
print("original matica = " + str(matica))

#reshape
reshape = matica.reshape(M,N)
print("reshape = " + str(reshape))
```
