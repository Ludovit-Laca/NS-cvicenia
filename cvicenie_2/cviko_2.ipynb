4.) *Neural network* - **AND function**

```
# X - input layer 
X = np.array([[0,0], [0,1], [1,0], [1,1]])

# Y - output layer 
Y = np.array([[0], [0], [0], [1]])
```

```
def akt_function(x):
  return 1/(1 + np.exp(-x))

# np.exp(1) is euler's number
```

```
def derivate_akt_function(x):
  return x * (1-x)

# euler's number^x' - euler's number^x
```

5.) Define weights for the hidden layer and the output layer

```
inputLayerSize = 2
hiddenLayerSize = 4
outputLayerSize = 1

weight_hidden_layer = np.random.uniform(low=0.0, high=1.0, size=(inputLayerSize, hiddenLayerSize))
weight_output_layer = np.random.uniform(low=0.0, high=1.0, size=(hiddenLayerSize, outputLayerSize))

print("Hidden layer weight = " + str(weight_hidden_layer) + "\n")
print("Output layer weight = " + str(weight_output_layer))
```

6.) Neural network training 

```
pole_error = []

for i in range(10000):
  hidden_layer = akt_function(np.matmul(X, weight_hidden_layer))
  output_layer = akt_function(np.matmul(hidden_layer, weight_output_layer))
  error = Y - output_layer

  #print("Hidden layer = " + str(hidden_layer) + "\n")
  #print("Output layer = " + str(output_layer) + "\n")
  #print("Error = " + str(error) + "\n")

  delta_Z = error*derivate_akt_function(output_layer)
  delta_H = delta_Z.dot(weight_output_layer.T)*derivate_akt_function(hidden_layer)
  weight_output_layer += hidden_layer.T.dot(delta_Z)
  weight_hidden_layer += X.T.dot(delta_H)

  pole_error.append(error[3])


print("Output layer = " + str(output_layer) + "\n")
print("Output layer weight = " + str(weight_output_layer) + "\n")
print("Hidden layer weight = " + str(weight_hidden_layer) + "\n")
print("Error = " + str(error) + "\n")

plt.plot(pole_error)
```
